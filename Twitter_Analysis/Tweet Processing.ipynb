```python
import math
import pandas as pd
import numpy as np
import csv
import sys, os
import string
import re
```

```python
dataloc = '/Volumes/Seagate Backup Plus Drive/PoliTweet/TwitterData/'
```

```python
files = os.listdir(dataloc); files
```

```python
temp = [re.sub('[-,\.,\[,\]]', ",", row) for row in files]; temp
```

```python
temp = [row.split(",") for row in temp]; temp
```

```python
newFiles = pd.DataFrame(temp)
```

```python
newFiles.columns = ['Del','Del2','Query','Candidate','Del3','Year','Month','Day','Hr','Min','Sec']
```

```python
newFiles.drop('Del', axis=1, inplace=True)
```

```python
newFiles.drop('Del2', axis=1, inplace=True)
```

```python
newFiles.drop('Del3', axis=1, inplace=True)
```

```python
HC_batch = newFiles[(newFiles['cleanCandidate'] == 'HillaryClinton') & (newFiles['Year'] == '2016') \
                    & (newFiles['Month'] == '06') & (newFiles['Day'] == '01') & (newFiles['Query'] == 'at')]
```

```python
# Grab Current 'Candidate' info for manipulation:
temp = newFiles.Candidate

# Split 'Candidate' column
temp = [row.split("#") for row in temp]
# Make new Clean Candidate column, without Hashtag appended

cleanCandidate = [row[0] for row in temp]
newFiles['cleanCandidate'] = cleanCandidate

# Extracting the HashTags from the 'Candidate' Column
hashtag = [row[1] if len(row) >= 2 else np.NaN for row in temp ]
newFiles['hashtag'] = hashtag
```

```python
newFiles.to_csv('masterFileList.csv', index = False)
```
