# Aggregating Polling Data from Real Clear Politics

```python
dataloc = '/Users/hopeemac/Desktop/pf7h/ElectionAnalytics/Data/'
```

```python
sampfile = 'Huffington Post/HuffPost-Democrat-2016-05-16.csv'
```

```python
pollDF = pd.DataFrame()
tempList = []
with open(dataloc+sampfile,'rt', errors = 'ignore') as f:
    reader = csv.reader(f)
    count = 0
    for row in reader:
        tempList.append(row)
        count += 1
    tempDF = pd.DataFrame(tempList[1:], columns = tempList[0])
    pollDF = pollDF.append(tempDF)
```
